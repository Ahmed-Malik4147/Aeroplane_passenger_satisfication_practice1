```python
df_business.drop(columns=["Class"], inplace=True)
```

```python
df_business.columns = df_business.columns.str.lower().str.replace(" ", "_")

df_business.columns
```

```python
df_business["age"] = pd.to_numeric(df_business["age"], errors="coerce")
```

```python
df_business["customer_type"] = df_business["customer_type"].str.lower()
df_business["satisfaction"] = df_business["satisfaction"].str.lower()
df_business["type_of_travel"] = df_business["type_of_travel"].str.lower()
df_business["gender"] = df_business["gender"].str.lower()
```

```python
df_business.isnull().sum()
```

```python
num_cols = df_business.select_dtypes(include=["float64", "int64"]).columns

for col in num_cols:
    df_business[col].fillna(df_business[col].median(), inplace=True)
```

```python
cat_cols = df_business.select_dtypes(include=["object"]).columns

for col in cat_cols:
    df_business[col].fillna(df_business[col].mode()[0], inplace=True)
```

```python
df_business.drop_duplicates(inplace=True)
```
